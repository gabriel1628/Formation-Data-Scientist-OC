# Fonctions

```python
# Scree plot
def display_scree_plot(pca):
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xticks(np.arange(0, pca.n_components_)+1)
    plt.xlabel("rang de l'axe d'inertie")
    plt.ylabel("pourcentage d'inertie")
    plt.title("Eboulis des valeurs propres")
    
# Cercles des corrélations
def display_circles(pcs, n_comp, pca, axis_ranks, labels=None, label_rotation=0, lims=None):
    for d1, d2 in axis_ranks: # On affiche les plans factoriels dans axis_ranks
        if d2 < n_comp:

            # initialisation de la figure
            fig, ax = plt.subplots(figsize=(7,6))

            # détermination des limites du graphique
            if lims is not None :
                xmin, xmax, ymin, ymax = lims
            elif pcs.shape[1] < 30 :
                xmin, xmax, ymin, ymax = -1, 1, -1, 1
            else :
                xmin, xmax, ymin, ymax = min(pcs[d1,:]), max(pcs[d1,:]), min(pcs[d2,:]), max(pcs[d2,:])

            # affichage des flèches
            # s'il y a plus de 30 flèches, on n'affiche pas le triangle à leur extrémité
            if pcs.shape[1] < 30 :
                plt.quiver(np.zeros(pcs.shape[1]), np.zeros(pcs.shape[1]),
                   pcs[d1,:], pcs[d2,:], 
                   angles='xy', scale_units='xy', scale=1, color="grey")
                # (voir la doc : https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html)
            else:
                lines = [[[0,0],[x,y]] for x,y in pcs[[d1,d2]].T]
                ax.add_collection(LineCollection(lines, axes=ax, alpha=.1, color='black'))

            # affichage des noms des variables  
            if labels is not None:  
                for i,(x, y) in enumerate(pcs[[d1,d2]].T):
                    if x >= xmin and x <= xmax and y >= ymin and y <= ymax :
                        plt.text(x, y, labels[i], fontsize='14', ha='center', va='center', 
                                 rotation=label_rotation, color="blue", alpha=0.5)

            # affichage du cercle
            circle = plt.Circle((0,0), 1, facecolor='none', edgecolor='b')
            plt.gca().add_artist(circle)

            # définition des limites du graphique
            plt.xlim(xmin, xmax)
            plt.ylim(ymin, ymax)

            # affichage des lignes horizontales et verticales
            plt.plot([-1, 1], [0, 0], color='grey', ls='--')
            plt.plot([0, 0], [-1, 1], color='grey', ls='--')

            # nom des axes, avec le pourcentage d'inertie expliqué
            plt.xlabel('F{} ({}%)'.format(d1+1, round(100*pca.explained_variance_ratio_[d1],1)))
            plt.ylabel('F{} ({}%)'.format(d2+1, round(100*pca.explained_variance_ratio_[d2],1)))

            plt.title("Cercle des corrélations (F{} et F{})".format(d1+1, d2+1))
```

## KMeans

```python
from sklearn.cluster import KMeans
```

```python
# préparation des données pour le clustering
X = df_rfm.copy()

# Centrage et réduction
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
```

### Analyse avec Yellowbrick

```python
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
sns.set_theme()
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
# Instantiate the clustering model and visualizer
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2,12))

visualizer.fit(X_scaled)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

By default, the scoring parameter `metric` is set to `distortion`, which computes the sum of squared distances from each point to its assigned center. However, two other metrics can also be used with the `KElbowVisualizer`, `silhouette` and `calinski_harabasz`. <br>
In the following, we'll use the `calinski_harabasz`.

```python
# Instantiate the clustering model and visualizer 
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2,12), metric='calinski_harabasz')

visualizer.fit(X_scaled)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

Une autre façon de quantifier à quel point un clustering répond à l'homogénéité et la séparation est de mesurer ce que l'on appelle le coefficient de silhouette.
<br>
Pour un point x donné, le coefficient de silhouette s(x) permet d'évaluer si ce point appartient au « bon » cluster : est-il proche des points du cluster auquel il appartient ? Est-il loin des autres points ?
<br>
Pour répondre à la première question, on calcule la distance moyenne de x à tous les autres points du cluster $C_k$ auquel il appartient :
$
a(x)=\frac{1}{|C_k|−1} \sum_{u\in C_k, u\neq x} d(u,x) \ .
$
<br>
Pour répondre à la deuxième, on calcule la plus petite valeur que pourrait prendre a(x), si x était assigné à un autre cluster :
$
b(x)=min_{l\neq k} \frac{1}{|Cl|} \sum_{u\in Cl} d(u,x) \ . 
$
<br>
Si x a été correctement assigné, alors a(x) < b(x). Le coefficient de silhouette est donné par 
$
s(x) = \frac{b(x)−a(x)}{max(a(x),b(x))} \ .
$
<br>
Il est donc compris entre -1 et 1, et d'autant plus proche de 1 que l'assignation de x à son cluster est satisfaisante. Pour évaluer un clustering, on peut calculer son coefficient de silhouette moyen.

```python
X_sample = X.sample(10000)
X_scaled_sample = scaler.transform(X_sample)
X_scaled_sample = pd.DataFrame(X_scaled_sample, index=X_sample.index, columns=X_sample.columns)
```

```python
n_clust = 3
# Instantiate the clustering model and visualizer
kmeans = KMeans(n_clust)
visualizer = SilhouetteVisualizer(kmeans)

visualizer.fit(X_scaled_sample)    # Fit the data to the visualizer
visualizer.poof()    # Draw/show/poof the data
```

Les clusters 0 et 2 sont beaucoup plus larges que les autres, ce qui signifie qu'ils contiennent plus d'échantillons. On peut voir aussi que le coefficient de silhouette est négatif pour certains échantillons, ce qui signifie qu'ils ont été attribués au mauvais cluster. La ligne en pointillé indique la valeur moyenne du coefficient de silhouette.

On peut calculer directement le coefficient de silhouette moyen avec `sklearn.metrics.silhouette_score`

```python
from sklearn.metrics import silhouette_score
labels = kmeans.labels_
silhouette_score(X_scaled_sample, labels, metric='euclidean')
```

### Visualisation sur les plans factoriels

```python
n_clust = 3

# Clustering par K-means
kmeans = KMeans(n_clusters=n_clust)
kmeans.fit(X_scaled)

# Récupération des clusters attribués à chaque individu
clusters = kmeans.labels_
np.unique(clusters)
```

```python
# Nombre de clients par cluster
pd.Series(clusters).value_counts()
```

```python
# PCA
n_comp = 3
pca = decomposition.PCA(n_components=n_comp).fit(X_scaled)
X_projected = pca.transform(X_scaled)

# Scree plot
display_scree_plot(pca)

# Cercle des corrélations
pcs = pca.components_
axis_ranks = [(0,1), (1,2)]
labels = X.columns
display_circles(pcs, n_comp, pca, axis_ranks, labels=labels)

# Affichage du clustering par projection des individus sur les plans factoriels
for (i,j) in axis_ranks:
    plt.figure()
    for cluster in np.unique(clusters):
        mask = clusters == cluster
        plt.scatter(X_projected[mask, i], X_projected[mask, j], label='Groupe {}'.format(cluster))
    plt.title('Projection des {} individus'.format(X_projected.shape[0]))
    plt.xlabel('F{}'.format(i+1))
    plt.ylabel('F{}'.format(j+1))
    plt.legend();
```

Visualisation 3D

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection='3d')
x = X_projected[:, 0]
y = X_projected[:, 1]
z = X_projected[:, 2]
ax.scatter(x, y, z, c=clusters, cmap='viridis')
ax.set_xlabel('F1')
ax.set_ylabel('F2')
ax.set_zlabel('F3');
```

### Visualisation dans l'espace des variables

```python
for (i, j) in ((0,1), (0,2), (1,2)):
    plt.figure()
    for cluster in np.unique(clusters):
        mask = clusters == cluster
        plt.scatter(X.iloc[mask, i], X.iloc[mask, j], label='Groupe {}'.format(cluster))
    #plt.xticks(rotation=30)
    plt.xlabel(X.columns[i])
    plt.ylabel(X.columns[j])
    plt.legend();
```

## Clustering hiérarchique

```python
from scipy.cluster.hierarchy import linkage, fcluster
```

```python
n_clust
```

```python
X_sample = X.sample(20000)
X_scaled_sample = scaler.transform(X_sample)
X_scaled_sample = pd.DataFrame(X_scaled_sample, index=X_sample.index, columns=X_sample.columns)
X_projected = pca.transform(X_scaled_sample)
```

```python
Z = linkage(X_scaled_sample, 'ward')
```

```python
# Coupage du dendrogramme en n clusters
clusters = fcluster(Z, n_clust, criterion='maxclust')
```
