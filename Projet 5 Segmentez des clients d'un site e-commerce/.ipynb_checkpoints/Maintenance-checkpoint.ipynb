```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
sns.set_color_codes('deep')

from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn import decomposition

plt.rcParams['figure.figsize'] = (8, 6)

meanprops = {'marker':'o',
             'markerfacecolor':'white', 
             'markeredgecolor':'black',
             'markeredgewidth': 1.5,
             'markersize':'10'}
```

# Clustering avec la segmentation RFMSA

```python
from sklearn.cluster import KMeans
```

```python
# Récence Fréquence Montant Score Articles par commande
df_rfmsa = pd.read_csv('df_rfmsa.csv', index_col=0)

# préparation des données pour le clustering
X_rfmsa = df_rfmsa.copy()

# Centrage et réduction
scaler = StandardScaler()
X_rfmsa_scaled = scaler.fit_transform(X_rfmsa)
X_rfmsa_scaled = pd.DataFrame(X_rfmsa_scaled, columns=X_rfmsa.columns, index=X_rfmsa.index)
```

```python
n_clust = 6

# Clustering par K-means
kmeans = KMeans(n_clusters=n_clust, random_state=12)
kmeans.fit(X_rfmsa_scaled)

# Récupération des clusters attribués à chaque individu
clusters = kmeans.labels_
np.unique(clusters)
```

Visualisation des clusters à l'aide de box plots

```python
df_rfmsa['cluster'] = clusters

for col in df_rfmsa.columns[:-1]:
    plt.figure()
    sns.boxplot(x=col, y='cluster', data=df_rfmsa, orient='h', showfliers=True, showmeans=True,
               meanprops=meanprops)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
date_ini = 130
days_range_init = date_ini #90
days_range = days_range_init
days_step_init = 10
days_step = days_step_init
n_clust = 6
threshold = .8

# On commence par sélectionner les commandes qui ont été effectuées il y a plus de
# date_ini jours
X_0 = X_rfmsa[X_rfmsa['recency'] > date_ini].copy()

# Model initial
model_init = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
model_init.fit(X_0)

ari = [] # liste qui va contenir les valeurs de ARI

# On va rajouter toutes les commandes qui ont eu lieu pendant days_range depuis 
# date_ini en faisant des pas de days_step et à chaque pas on va comparer le
# clustering obtenu par l'ancien modèle (kmeans_0) avec celui obtenu par le
# nouveau modèle (kmeans_1) en utilisant l'ARI. 
# Si la valeur de l'ARI descend en-dessous de threshold, on réinitialise la
# valeur de date_ini et on recommence.
n_days = np.arange(0, days_range+1, days_step)
for n in n_days:
    # On rajoute les commandes qui ont été faites pendant n jours après date_ini
    X_1 = X_rfmsa[X_rfmsa['recency'] > date_ini - n].copy()

    # Clustering avec le modèle initial
    clusters_init = model_init.predict(X_1)

    # On entraîne un nouveau modèle.
    model_new = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
    model_new.fit(X_1)
    # Récupération des clusters attribués à chaque individu
    clusters_new = model_new['kmeans'].labels_

    ari += [adjusted_rand_score(labels_true=clusters_new, labels_pred=clusters_init)]

ari = np.array(ari)
```

```python
# On trace les valeurs d'ARI
n0 = date_ini - days_range
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(n_days[::-1]+n0, ari, 'o-')
ax.set_xlabel('Recency')
ax.set_ylabel('ARI')
ax.plot(n_days, np.zeros(n_days.shape)+threshold, '--')
ax.invert_xaxis()
plt.show()
```

```python
date_ini = 150
days_range_init = 50
days_range = days_range_init
days_step_init = 5
days_step = days_step_init
n_clust = 6
threshold = .8

for i in range(6):
    # On commence par sélectionner les commandes qui ont été effectuées il y a plus de
    # date_ini jours
    X_0 = df_rfmsa[df_rfmsa['recency'] > date_ini].copy()
    
    # Model initial
    model_init = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
    model_init.fit(X_0)

    ari = [] # liste qui va contenir les valeurs de ARI
    
    # On va rajouter toutes les commandes qui ont eu lieu pendant days_range depuis 
    # date_ini en faisant des pas de days_step et à chaque pas on va comparer le
    # clustering obtenu par l'ancien modèle (kmeans_0) avec celui obtenu par le
    # nouveau modèle (kmeans_1) en utilisant l'ARI. 
    # Si la valeur de l'ARI descend en-dessous de threshold, on réinitialise la
    # valeur de date_ini et on recommence.
    n_days = np.arange(0, days_range, days_step)
    for n in n_days:
        # On rajoute les commandes qui ont été faites pendant n jours après date_ini
        X_1 = df_rfmsa[df_rfmsa['recency'] > date_ini - n].copy()

        # Clustering avec le modèle initial
        clusters_init = model_init.predict(X_1)

        # On entraîne un nouveau modèle.
        model_new = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
        model_new.fit(X_1)
        # Récupération des clusters attribués à chaque individu
        clusters_new = model_new['kmeans'].labels_

        ari += [adjusted_rand_score(labels_true=clusters_new, labels_pred=clusters_init)]
        
    print('# of days before last order :', date_ini, 
          '  range of study (in days) :', days_range)
    ari = np.array(ari)
    
    # On trace les valeurs d'ARI
    plt.figure(figsize=(8,6))
    plt.plot(n_days, ari)
    plt.xlabel('# days')
    plt.ylabel('ARI')
    plt.plot(n_days, np.zeros(n_days.shape)+threshold, '--')
    plt.show()
    
    # On regarde s'il y a une valeur de ARI inférieure à threshold
    if (ari < threshold).sum() > 0:
        delta = n_days[ari < threshold][0]
        date_ini -= delta
        days_range = days_range_init
        days_step = days_step_init
    # Si ce n'est pas le cas, on double la valeur de days_range si ça ne va pas 
    # au-delà de date_ini
    elif (date_ini - 2*days_range) >= 0:
        days_range *= 2
        days_step *= 2
    # Si on a déjà parcouru toutes les commandes et qu'il n'y a pas de valeur d'ARI
    # en dessous du seuil, on arrête la boucle
    elif days_range == date_ini:
        break
    # Sinon, on parcourt toutes les commandes restantes en posant days_range = date_ini
    else :
        days_range = date_ini
```
