# Segmentation RFM

## Premier essai

### Récence

```python
df.shape
```

```python
df_ = df.sort_values(by='order_purchase_timestamp')
print(df_.shape)
df_.head()
```

```python
# One ne garde que la dernière commande de chaque client
df_last = df_.drop_duplicates(subset='customer_unique_id', keep='last')
print(df_last.shape)
df_last.head()
```

```python
# On a autant de lignes de que de clients différents
df['customer_unique_id'].unique().size
```

On vérifie qu'on a bien retenu les dates des dernières commandes

```python
# On prend les clients qui ont effectués plus de 10 articles
counts = df['customer_unique_id'].value_counts()
counts = counts[counts > 10]
customers = counts.index

last_orders = pd.Series()
last_orders_2 = pd.Series()
for customer in customers:
    # date de la dernière commande dans df
    last_order = df[df['customer_unique_id'] == customer].sort_values(by='order_purchase_timestamp').iloc[[-1]]
    last_order = last_order['order_purchase_timestamp']
    last_orders = pd.concat([last_orders, last_order])
    # date de la commande correspondante dans df_last
    last_order_2 = df_last.loc[df_last['customer_unique_id'] == customer, 'order_purchase_timestamp']
    last_orders_2 = pd.concat([last_orders_2, last_order_2])

# S'il n'y a aucune ligne c'est que les dates sont les mêmes dans les deux séries
pd.DataFrame({'df': last_orders.values, 'df_last': last_orders_2.values}).T.drop_duplicates(keep=False)
```

```python
recence = df_last[['customer_unique_id', 'order_purchase_timestamp']]
recence.head()
```

On va définir la récence comme le nombre de jours écoulés depuis la dernière commande

```python
s = recence['order_purchase_timestamp']
recence['recency'] = s.max() - s
recence['recency'] = recence['recency'].dt.days
recence = recence.drop(columns=['order_purchase_timestamp'])
recence.head()
```

### Fréquence

On va définir la fréquence comme étant le nombre de fois qu'un client a fait une commande (pour rappel, une commande peut contenir plusieurs articles, le nombre d'articles ne correspond donc pas forcément au nombre de commandes)

```python
df_ = df.drop_duplicates(subset='order_id')
df_.shape
```

```python
frequence = df_['customer_unique_id'].value_counts().reset_index()
frequence.columns = ['customer_unique_id', 'frequency']
frequence
```

### Montant

Pour le montant, on va prendre le prix moyen des commandes

```python
# nombre de commandes des différents clients
n_orders = df.drop_duplicates(subset='order_id', keep='first')['customer_unique_id'].value_counts()
n_orders
```

```python
sum_price = df.groupby(by='customer_unique_id').sum()['price']
sum_price
```

```python
montant = sum_price / n_orders
montant = montant.reset_index()
montant.columns = ['customer_unique_id', 'mean_spending']
montant
```

```python
df_rfm = pd.merge(recence, frequence)
df_rfm = df_rfm.merge(montant)
df_rfm.set_index(keys='customer_unique_id', inplace=True)
df_rfm.head()
```

```python
print(recence.shape)
print(frequence.shape)
print(montant.shape)
print(df_rfm.shape)
```

```python
df_rfm.to_csv('df_rfm.csv')
```
