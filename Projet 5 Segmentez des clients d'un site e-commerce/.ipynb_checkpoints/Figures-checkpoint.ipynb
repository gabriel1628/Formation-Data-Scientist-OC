# Segmentation RFM

```python
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
```

```python
df_rfm = pd.read_csv('df_rfm.csv', index_col=0)
df_rfm.head()
```

```python
n_clust = 4
X_rfm = df_rfm.copy()

# On définit une pipeline dans laquelle on va normaliser les data puis appliquer le clustering avec KMeans
model = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
model.fit(X_rfm)

# Récupération des clusters attribués à chaque individu
clusters = model['kmeans'].labels_
np.unique(clusters)
```

```python
#df_rfm.columns = ['Récence (jours)', 'Fréquence', 'Montant']

df_rfm['cluster'] = clusters

for col in df_rfm.columns[:-1]:
    plt.figure(figsize=(8,6))
    sns.boxplot(x=col, y='cluster', data=df_rfm, orient='h', showfliers=True, showmeans=True,
               meanprops=meanprops)
    if col == 'Montant':
        plt.xlim(-500, 5000)
```

```python
df_rfm['cluster'].value_counts().sort_index()
```

# Segmentation RFMSA

```python
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
```

```python
df_rfmsa = pd.read_csv('df_rfmsa.csv', index_col=0)
df_rfmsa.head()

X_rfmsa = df_rfmsa.copy()
```

```python
n_clust = 6
X_rfm = df_rfmsa.copy()

# On définit une pipeline dans laquelle on va normaliser les data puis appliquer le clustering avec KMeans
model = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
model.fit(X_rfm)

# Récupération des clusters attribués à chaque individu
clusters = model['kmeans'].labels_
```

```python
def switch(a, i, j):
    store = 2*(i+j) + 5
    a[a == i] = store
    a[a == j] = i
    a[a == store] = j
```

```python
switch(clusters, 0, 1)
switch(clusters, 2, 5)
switch(clusters, 3, 4)
```

```python
df_rfmsa['cluster'] = clusters

df_rfmsa.columns = ['Récence (jours)', 'Fréquence', 'Montant', 'Score', 'Nb d\'Articles', 'cluster']

for col in df_rfmsa.columns[:-1]:
    plt.figure(figsize=(8,6))
    sns.boxplot(x=col, y='cluster', data=df_rfmsa, orient='h', showfliers=True, showmeans=True,
               meanprops=meanprops)
    if col == 'Montant':
        plt.xlim(-500, 5000)
```

```python
df_rfmsa['cluster'].value_counts().sort_index()
```

# ARI

```python
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
recence_init = 130 # date/récence à laquelle on initialise le modèle initial
days_range = recence_init # nombre de jours où on va calculer l'ARI
days_step = 10 # pas entre deux mesures d'ARI
threshold = .8

# On commence par sélectionner les commandes qui ont été effectuées il y a plus de
# recence_init jours
X_0 = X_rfmsa[X_rfmsa['recency'] > recence_init].copy()

# Model initial
model_init = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
model_init.fit(X_0)

ari = [] # liste qui va contenir les valeurs d'ARI

# On va rajouter toutes les commandes qui ont eu lieu pendant days_range depuis 
# recence_init en faisant des pas de days_step et à chaque pas on va comparer le
# clustering obtenu par l'ancien modèle (model_init) avec celui obtenu par le
# nouveau modèle (model_new) en utilisant l'ARI.
n_days = np.arange(0, days_range+1, days_step)
for n in n_days:
    # On rajoute les commandes qui ont été faites pendant n jours après recence_init
    X_1 = X_rfmsa[X_rfmsa['recency'] > recence_init - n].copy()

    # Clustering avec le modèle initial
    clusters_init = model_init.predict(X_1)

    # On entraîne un nouveau modèle.
    model_new = make_pipeline(StandardScaler(), KMeans(n_clusters=n_clust, random_state=12))
    model_new.fit(X_1)
    # Récupération des clusters attribués à chaque individu
    clusters_new = model_new['kmeans'].labels_

    ari += [adjusted_rand_score(labels_true=clusters_new, labels_pred=clusters_init)]

ari = np.array(ari) # conversion en ndarray
```

```python
# On trace les valeurs d'ARI
delta = recence_init - days_range
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(n_days[::-1]+delta, np.zeros(n_days.shape)+threshold, '--', color='#dd8452')
ax.set_xlabel('# jours après initialisation')
ax.set_ylabel('ARI')

ax2 = ax.twiny()
ax2.plot(n_days[::-1]+delta, ari, 'o-')
ax2.set_xlabel('Récence')
ax2.invert_xaxis()

plt.show()
```

# Projection ACP

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_rfm = pd.read_csv('df_rfm.csv', index_col=0)
X_rfm = df_rfm.copy()
X_rfm_scaled = StandardScaler().fit_transform(X_rfm)

# Clustering par K-means
n_clust = 4
kmeans = KMeans(n_clusters=n_clust, random_state=5)
kmeans.fit(X_rfm_scaled)

# Récupération des clusters attribués à chaque individu
clusters = kmeans.labels_
np.unique(clusters)
```

```python
# ACP
n_comp = 2 #nombre de composantes
pca = decomposition.PCA(n_components=n_comp).fit(X_rfm_scaled)
X_rfm_projected = pca.transform(X_rfm_scaled)
```

```python
pcs = pca.components_
axis_ranks = [(0,1)]
labels = X_rfm.columns
```

```python
# Affichage du clustering par projection des individus sur les plans factoriels
for (i,j) in axis_ranks:
    plt.figure(figsize=(12,8))
    for cluster in np.unique(clusters):
        mask = clusters == cluster
        plt.scatter(X_rfm_projected[mask, i], X_rfm_projected[mask, j], label='Cluster {}'.format(cluster))
    plt.title('Projection des individus')
    plt.xlabel('F{}'.format(i+1))
    plt.ylabel('F{}'.format(j+1))
    plt.xlim(-10,10)
    plt.ylim(-5,10)
    plt.xticks([])
    plt.yticks([])
    plt.legend();
```

```python
# Affichage du clustering par projection des individus sur les plans factoriels et de la projection
# des variables mises à léchelle selon les limites du graphe
for (i,j) in axis_ranks:
    fig, ax = plt.subplots(figsize=(12,8))
    for cluster in np.unique(clusters):
        mask = clusters == cluster
        ax.scatter(X_rfm_projected[mask, i], X_rfm_projected[mask, j], label='Cluster {}'.format(cluster))
        
    # Flèches
    xmax = 10 #ax.get_xlim()[1]
    xmin = -10
    ymax = 10 #ax.get_ylim()[1]
    ymin = -5
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    plt.quiver(np.zeros(pcs.shape[1]), np.zeros(pcs.shape[1]),
    pcs[i,:]*xmax, pcs[j,:]*ymax, 
    angles='xy', scale_units='xy', scale=1, color="grey")
    
    plt.title('Projection des individus')
    plt.xlabel('F{}'.format(i+1))
    plt.ylabel('F{}'.format(j+1))
    plt.xticks([])
    plt.yticks([])
    plt.legend();
```
