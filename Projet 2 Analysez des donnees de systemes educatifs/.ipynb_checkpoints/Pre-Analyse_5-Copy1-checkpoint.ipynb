Accès à internet

Prévision des population cibles

```python
liste = ["Projection: Population age 15-19 in thousands by highest level of educational attainment. Upper Secondary. Total",
         "Projection: Population age 20-24 in thousands by highest level of educational attainment. Upper Secondary. Total",
         "Projection: Population age 15-19 in thousands by highest level of educational attainment. Post Secondary. Total",
         "Projection: Population age 20-24 in thousands by highest level of educational attainment. Post Secondary. Total",
        ]
condition = data5["Indicator Name"] == liste[0]
for val in liste[1:]:
    cond = data5["Indicator Name"] == val
    condition = condition | cond
```

## On crée un dataframe pour les scores

```python
data_scores = data[["Country Code", "Country Name"]]
data_scores.head()
```

## Création du score INTERNET

```python
# On fait le score en divisant par la valeur max
data_scores["SCORE_INTERNET"] = data["Internet users"] / data["Internet users"].max()
data_scores.head()
```

## Création du score INCOME

```python
# On fait le score en divisant par la valeur max
data_scores["SCORE_INCOME"] = data["GNI per capita"] / data["GNI per capita"].max()
data_scores.head()
```

## Création du score POPULATION

```python
# On fait le score en divisant par la valeur max
data_scores["SCORE_POPULATION"] = data["Population"] / data_pop["Population"].max()
data_scores.head()
```

## Création du score PROJECTION

```python
data.head()
```

```python
# On fait le score en divisant par la valeur max
data_scores["SCORE_PROJECTION"] = data["Projection"] / data["Projection"].max()
data_scores.head()
```

```python
# On vérifie qu'il n'y a pas de nan
data_scores.isna().sum()
```

### Regardons le classement

```python
data_scores["SCORE_TOTAL"] = (data_scores["SCORE_INCOME"]
                              + data_scores["SCORE_INTERNET"]
                              + data_scores["SCORE_POPULATION"]
                              + data_scores["SCORE_PROJECTION"]) / 4
data_scores.head()
```

```python
# Regardons les 10 premiers
data_scores.sort_values("SCORE_TOTAL", ascending=False)[:10]
```

```python
data_scores.columns[:1:-1]
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Make the PairGrid
g = sns.PairGrid(data_scores.sort_values("SCORE_TOTAL", ascending=False),
                 x_vars=data_scores.columns[:1:-1], y_vars=["Country Name"],
                 height=10, aspect=.25)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=10, orient="h", jitter=False,
      palette="flare_r", linewidth=1, edgecolor="w")

# Draw a bar plot using the barplot function
#g.map(sns.barplot, palette="flare_r")

# Use the same x axis limits on all columns and add better labels
g.set(xlim=(-0.05, 1.05), xlabel="Score", ylabel="")

# Use semantically meaningful titles for the columns
titles = ["Total score", "Projection score", "Population score",
          "Income score", "Internet score"]

for ax, title in zip(g.axes.flat, titles):

    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True)
```
